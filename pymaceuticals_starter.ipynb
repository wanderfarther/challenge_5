### Analysis

- The correlation between the weight of a mouse and the average tumor size was .84, which means these two factors have a positive correlation to each other. The larger the mouse was or became, the large the tumar became as well. 
- Infubinol had an outlier in the volume of the tumors, which can be seen in the box plot. This would be something to look into if we moved into further treatments with Infubinol. 
- After cleaning the data, there was 13 rows of duplicated data. Without removing these rows, the data could have been skewed towards Propriva.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_study_df = pd.merge(mouse_metadata, study_results, how ="outer")

# Display the data table for preview
mouse_study_df
```

```python
# Checking the number of mice.
len(mouse_study_df['Mouse ID'].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_rows = mouse_study_df[mouse_study_df.duplicated(['Mouse ID','Timepoint'])]
duplicate_rows
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_data = mouse_study_df[mouse_study_df.duplicated(['Mouse ID','Timepoint'], keep=False)]
dup_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_study_clean_df = mouse_study_df.drop(mouse_study_df[mouse_study_df['Mouse ID'] == 'g989'].index)
mouse_study_clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
len(mouse_study_clean_df['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen


# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame

drug_regimen_grouped = mouse_study_clean_df.groupby('Drug Regimen')
group_mean = drug_regimen_grouped['Tumor Volume (mm3)'].mean()
group_median = drug_regimen_grouped['Tumor Volume (mm3)'].median()
group_var = drug_regimen_grouped['Tumor Volume (mm3)'].var()
group_std = drug_regimen_grouped['Tumor Volume (mm3)'].std()
group_sem = drug_regimen_grouped['Tumor Volume (mm3)'].sem()

averages_df = pd.DataFrame(group_mean)
averages_df = averages_df.rename(columns={'Tumor Volume (mm3)':'Mean Tumor Volume'})
averages_df['Median Tumor Volume'] = group_median
averages_df['Tumor Volume Variance'] = group_var
averages_df['Tumor Volume Std. Dev.'] = group_std
averages_df['Tumor Volume Std. Err.'] = group_sem

averages_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
sum_stat_table = mouse_study_clean_df.groupby('Drug Regimen').agg({'Tumor Volume (mm3)':['mean','median','var','std','sem']})
sum_stat_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
mouse_grouped = mouse_study_clean_df.groupby('Mouse ID')['Timepoint'].max()
mouse_grouped

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
new_clean_mouse_study = pd.merge(mouse_study_clean_df, mouse_grouped, on=('Mouse ID','Timepoint'))
new_clean_mouse_study
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
for drug in drug_list:
    drug_tumor_vol = new_clean_mouse_study.loc[new_clean_mouse_study['Drug Regimen'] == drug, 'Tumor Volume (mm3)']
    
    # add subset 
    tumor_vol_data.append(drug_tumor_vol)
    quartile = drug_tumor_vol.quantile([0.25, 0.75])
    lowerq = quartile[0.25]
    upperq = quartile[0.75]
    iqr = upperq - lowerq
    lowerb = (lowerq - 1.5 * iqr)
    upperb = (upperq + 1.5 * iqr)

    # Determine outliers using upper and lower bounds
    outliers = drug_tumor_vol.loc[(drug_tumor_vol < lowerb) | (drug_tumor_vol > upperb)]

    print(f"{drug}'s potential outliers: {outliers}")
    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(tumor_vol_data, sym="ro", labels = drug_list)
plt.ylabel('Final Tumor Volume (mm3)')
plt.tight_layout()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
mouse_l509 = mouse_study_clean_df[mouse_study_clean_df['Mouse ID'] == 'l509']
x_value = mouse_l509['Timepoint']
y_value = mouse_l509['Tumor Volume (mm3)']
plt.plot(x_value, y_value)
plt.title('Capomulin treatment of mouse L509')
plt.ylabel('Tumor Volume (mm3)')
plt.xlabel('Timepoint (days)')
plt.tight_layout()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capolumin_only = mouse_study_clean_df.loc[mouse_study_clean_df['Drug Regimen'] == 'Capomulin']
cap_vol = capolumin_only.groupby('Mouse ID').mean(['Tumor Volume(mm3)'])
#cap_vol
x_values = cap_vol['Weight (g)']
y_values = cap_vol['Tumor Volume (mm3)']
plt.scatter(x_values, y_values)
plt.ylabel('Average Tumor Volume (mm3)')
plt.xlabel('Weight (g)')
plt.tight_layout()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
cor_coefficent, p_value = st.pearsonr(x_values, y_values)
print(f"The correlation between mouse weight and the average tumor volume is {round(cor_coefficent,2)}.")

#linear regression
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
plt.plot(x_values, (slope*x_values + intercept), color="red")
plt.scatter(x_values, y_values)
plt.ylabel('Average Tumor Volume (mm3)')
plt.xlabel('Weight (g)')
plt.tight_layout()
```
